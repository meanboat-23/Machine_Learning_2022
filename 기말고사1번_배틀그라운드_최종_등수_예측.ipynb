```python
train = pd.read_csv('/kaggle/input/2022-ml-finalexam-p1/train.csv')
test = pd.read_csv('/kaggle/input/2022-ml-finalexam-p1/test.csv')
sample = pd.read_csv('/kaggle/input/2022-ml-finalexam-p1/sample_submission.csv')
```

train데이터와 test데이터, 제출할 파일을 읽어 오기 위해서 pd.read_csv함수를 사용하였습니다.

```python
train
```

```python
X_train = train.drop(['Id','groupId','matchId','winPlacePerc'],axis=1)
X_test = test.drop(['Id','groupId','matchId'],axis=1)
y_train = train['winPlacePerc']
```

train데이터와 test데이터를 살펴보니 'Id','groupId','matchId'라는 부분은 모델의 학습에 필요하지 않을 것 같아서 제거해 주기 위해서 drop 함수를 사용하였습니다.

```python
X_train.info()
X_test.info()
```

train데이터와 test데이터의 각각의 데이터형식과 null값의 개수를 파악하기 위해서 info함수를 사용하였습니다.

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor(random_state=42).fit(X_train,y_train)
y_test = model.predict(X_test)
```

여러가지의 분류기중 최근 가장 좋은 성능을 보이고 있다는 lightgbm을 사용하기 위해서 해당 모델을 호출하여 학습하고 예측을 진행했습니다.

```python
sample['winPlacePerc'] = y_test
sample.to_csv('submit.csv',index=False)
```

예측한 값을 csv파일로 옮겨서 제출하기 위해서 to_csv함수를 사용했습니다.
